## Data Cleaning

The Client Want a Clean Data With Column (Do_Not_Contact) = No

```python
# Importing The Libarary
import pandas as pd
```

```python
#To Read File
df = pd.read_excel("/content/Customer Call List.xlsx")
df
```

Row 20 Contains Duplicates

```python
# To remove Duplicate
df.drop_duplicates()
```

```python
# Droping The Unwanted Column
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^0-9A-Za_z]'," ")
df['Phone_Number'] = df['Phone_Number'].str.replace('N a','')
df['Phone_Number'] = df['Phone_Number'].str.replace(' ',"-")
df
```

```python
# Replacing The value In Paying Customer Column
df['Paying Customer'] = df['Paying Customer'].str.replace('N/a','')
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df
```

```python
# Replacing The value In Do_Not_Contact Column
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df
```

```python
# Replacing The value In Address Column
df['Address'] = df['Address'].str.replace('N/a','')
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('123./')
df
```

```python
df.dropna(inplace= True)
df
```

```python
# Spliting The Column
df[["Street_Address", "State", "Zip_Code"]] = df['Address'].str.split(',',2, expand= True)
df
```

```python
New_Data = df[df['Do_Not_Contact'] != 'Y']
New_Data
```

```python
New_Data.drop(columns='Address', inplace= True)
New_Data
```
